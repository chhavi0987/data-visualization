```python
# prompt:  do the visualization code of /content/PMC Hospital Infrastructure.csv

!apt-get update
!apt-get install -y fonts-indic
!fc-cache -fv
!rm -rf /root/.cache/matplotlib/
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Load the dataset
df = pd.read_csv('/content/PMC Hospital Infrastructure.csv')

# Display the first few rows and information about the dataframe
print(df.head())
print(df.info())

# Clean up column names (remove leading/trailing spaces and special characters)
df.columns = df.columns.str.strip().str.replace(r'[^\w\s]', '', regex=True)
df.columns = df.columns.str.replace(' ', '_')

# Ensure column names are valid identifiers (optional but good practice)
df.columns = df.columns.str.replace('-', '_').str.replace('(', '').str.replace(')', '')

# Let's check the cleaned column names
print(df.columns)

# Assuming the dataset has columns like 'Hospital', 'Beds', 'Doctors', 'Nurses', etc.
# We need to know the exact column names from your file.
# For demonstration, let's assume we want to visualize 'Beds' per 'Hospital_Name'

# Replace 'Hospital_Name' and 'Total_Available_Beds' with the actual column names from your CSV
# You can identify these from the `print(df.columns)` output.
# Example: If your file has columns named 'Hospital Name' and 'No. of Beds',
# the cleaned names would likely be 'Hospital_Name' and 'No_of_Beds'.

# Example Visualization 1: Bar plot of the number of beds per hospital
# Replace 'Hospital_Name' and 'Total_Available_Beds' with the actual column names
hospital_column = 'Hospital_Name' # Replace with the actual hospital name column
beds_column = 'Total_Available_Beds' # Replace with the actual beds column

# Check if the required columns exist in the DataFrame
if hospital_column in df.columns and beds_column in df.columns:
    # Sort data for better visualization
    df_sorted_beds = df.sort_values(by=beds_column, ascending=False)

    plt.figure(figsize=(12, 8))
    sns.barplot(x=beds_column, y=hospital_column, data=df_sorted_beds)
    plt.title('Number of Beds per Hospital')
    plt.xlabel('Number of Beds')
    plt.ylabel('Hospital Name')
    plt.tight_layout()
    plt.show()
else:
    print(f"Error: Columns '{hospital_column}' or '{beds_column}' not found in the DataFrame.")
    print("Please check the column names in your CSV file and update the code accordingly.")
    print("Available columns:", df.columns.tolist())


# Example Visualization 2: Scatter plot comparing two numerical columns
# Replace 'Total_Available_Beds' and 'Total_Doctors' with actual numerical column names
doctors_column = 'Total_Doctors' # Replace with the actual doctors column name

if beds_column in df.columns and doctors_column in df.columns:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=beds_column, y=doctors_column, data=df)
    plt.title('Beds vs. Doctors per Hospital')
    plt.xlabel('Total Available Beds')
    plt.ylabel('Total Doctors')
    plt.tight_layout()
    plt.show()
else:
    print(f"Error: Columns '{beds_column}' or '{doctors_column}' not found in the DataFrame.")
    print("Please check the column names in your CSV file and update the code accordingly.")
    print("Available columns:", df.columns.tolist())


# Example Visualization 3: Distribution of a numerical column (e.g., Beds)
if beds_column in df.columns:
    plt.figure(figsize=(8, 5))
    sns.histplot(df[beds_column], kde=True)
    plt.title('Distribution of Total Available Beds')
    plt.xlabel('Number of Beds')
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()
else:
    print(f"Error: Column '{beds_column}' not found in the DataFrame.")
    print("Please check the column names in your CSV file and update the code accordingly.")
    print("Available columns:", df.columns.tolist())

# Add more visualizations based on available columns in your CSV.
# For example, if you have a column for 'Number of Nurses':
# nurses_column = 'Total_Nurses' # Replace with the actual nurses column name
# if nurses_column in df.columns:
#     plt.figure(figsize=(8, 5))
#     sns.histplot(df[nurses_column], kde=True)
#     plt.title('Distribution of Total Nurses')
#     plt.xlabel('Number of Nurses')
#     plt.ylabel('Frequency')
#     plt.tight_layout()
#     plt.show()

# Remember to replace the placeholder column names ('Hospital_Name', 'Total_Available_Beds', 'Total_Doctors', etc.)
# with the actual column names present in your '/content/PMC Hospital Infrastructure.csv' file.
# The `print(df.columns)` output after cleaning will show you the correct names to use.
```
